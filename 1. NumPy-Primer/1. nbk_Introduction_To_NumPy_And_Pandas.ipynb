# Introduction to NumPy and Pandas
* <b>NumPy</b> and <b>Pandas</b> are the two critical <b>Python Libraries</b> that help structure data in <b>Arrays</b> and <b>DataFrames</b> and contain built-in functions for <b>Data Analysis</b>.

# What is Pandas?
* <b>Pandas</b> is the most widely used <b>Python Library</b> for <b>Data Analysis</b>.
* <b>Pandas</b> contains functions for accessing, aggregating, joining and analyzing data.
* The most commonly used <b>Data Structure</b> in <b>Pandas</b> is the <b>DataFrame</b>, which is the <b>Tabular Data Structure</b> that is considered analogous to <b>SQL Tables</b> or <b>Excel Worksheets</b>. Just about anything that can be done in either <b>SQL</b> or <b>Excel</b> can be done in <b>Pandas</b>.

# What is NumPy?
* <b>NumPy</b> stands for <b>Numerical Python</b>.
* <b>NumPy</b> is an <b>Open-Source Python Library</b> that is the universal standard for working with numerical data in <b>Python</b>, and, forms the foundation of other <b>Python Libraries</b>, like - <b>Pandas</b>.
* <b>Pandas DataFrames</b> are built on <b>NumPy Arrays</b> and can leverage <b>NumPy</b> functions.
* In the below image of a <b>Pandas DataFrame</b>, it can be seen that all the <b>Rows</b> of any of the <b>Columns</b> in the <b>Pandas DataFrame</b> are stored as a <b>NumPy Array</b> -

* <b>Pandas</b> merely adds wrappers like - <b>Indices</b> and <b>Column Titles</b> to make working with those <b>NumPy Arrays</b> much more convenient.
* So, it is really difficult to work with <b>Pandas</b> without knowing some <b>NumPy</b>.
